```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
df_train = pd.read_csv("train_KNN.csv")
df_test = pd.read_csv("test_KNN.csv")

df_train.head()
```

```python
df_train["Age"] = df_train["Age"].fillna(df_train["Age"].median())
df_test["Age"] = df_test["Age"].fillna(df_test["Age"].median())

df_train = df_train.dropna(subset=["Embarked"])
df_test = df_test.dropna(subset=["Fare"])

df_train = df_train.drop(["Cabin", "Name", "Ticket", "PassengerId"], axis=1)
df_test = df_test.drop(["Cabin", "Name", "Ticket", "PassengerId"], axis=1)
```

```python
df_train = pd.get_dummies(df_train, drop_first=True)
df_test = pd.get_dummies(df_test, drop_first=True)
```

```python
X_train = df_train.drop("Survived", axis=1)
y_train = df_train["Survived"]

X_test = df_test.drop("Survived", axis=1)
y_test = df_test["Survived"]
```

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

knn_accuracy = accuracy_score(y_test, y_pred)
print("KNN Accuracy:", knn_accuracy)
```

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)

lr_accuracy = accuracy_score(y_test, y_pred_lr)
print("Logistic Regression Accuracy:", lr_accuracy)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)

dt_accuracy = accuracy_score(y_test, y_pred_dt)
print("Decision Tree Accuracy:", dt_accuracy)
```

```python
print("KNN Accuracy:", knn_accuracy)
print("Logistic Regression Accuracy:", lr_accuracy)
print("Decision Tree Accuracy:", dt_accuracy)
```

## Findings and Interpretation

In this experiment, three machine learning algorithms were applied to the Titanic dataset to predict passenger survival.

The K-Nearest Neighbors (KNN) model achieved an accuracy of approximately 81%, showing moderate performance. Logistic Regression performed the best with an accuracy of around 94%, indicating that it is the most suitable model for this dataset. The Decision Tree model achieved around 79% accuracy, which is slightly lower and may be affected by overfitting.

From the comparison, it is clear that Logistic Regression provides the most reliable results, while KNN and Decision Tree show comparatively lower performance.

This experiment demonstrates that different machine learning models produce different results, and choosing the right model is important for achieving high accuracy.
